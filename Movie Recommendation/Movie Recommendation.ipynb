```python
import pandas as pd 
```

```python
rating = pd.read_csv(r"C:\Users\MY LAPTOP\OneDrive\Desktop\Machine_Learning\Movie Recommendation\dataset\ratings.csv")
movies  = pd.read_csv(r"C:\Users\MY LAPTOP\OneDrive\Desktop\Machine_Learning\Movie Recommendation\dataset\movies.csv")
```

```python
rating.head()
```

```python
movies.head()
```

```python
ratings= pd.merge(movies, rating)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index=['userId'], columns=['title'], values='rating')
user_ratings.head()
```

```python
user_ratings = user_ratings.dropna(thresh=10, axis=1).fillna(0)
user_ratings.head()
```

```python
item_similarity = user_ratings.corr(method='pearson')
item_similarity.head()
```

```python
def get_similar_movies(movie_name, user_rating):
    similar_score = item_similarity[movie_name]*(user_rating - 2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score

get_similar_movies("'burbs, The (1989)", 4)
```
